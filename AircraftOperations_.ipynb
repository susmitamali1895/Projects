Identify numerical and categorical columns to apply appropriate missing value imputation strategies.


**Reasoning**:
To identify numerical and categorical columns, I will use `df.select_dtypes` to separate columns based on their data types and then print the resulting lists of column names.



```python
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

## Impute Numerical Missing Values

### Subtask:
Fill missing values in numerical columns with their respective mean values.


**Reasoning**:
To impute missing numerical values, I will iterate through each numerical column, calculate its mean, and then fill the NaN values with that mean. Afterwards, I will display the DataFrame's info and head to verify the changes.



```python
for col in numerical_cols:
    if df[col].isnull().any():
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)

df.info()
df.head()
```

## Impute Categorical Missing Values

### Subtask:
Fill missing values in categorical (object/string) columns with the string 'NA'.


**Reasoning**:
To impute missing categorical values, I will iterate through each categorical column and fill NaN values with the string 'NA'. Afterwards, I will display the DataFrame's info and head to verify the changes.



```python
for col in categorical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna('NA')

df.info()
df.head()
```
