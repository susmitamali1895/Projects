Pandas is an open-source software library for the Python programming language, primarily used for data manipulation and analysis. It provides powerful and flexible data structures designed to make working with "relational" or "labeled" data, such as tabular data (like spreadsheets or SQL tables) and time series, both easy and intuitive.


Matplolib
Matplotlib is an open-source, comprehensive plotting library for the Python programming language. Its primary function is to enable the creation of static, animated, and interactive visualizations in Python.

Pandas uses Matplotlib (and optionally Seaborn) as a backend to create a variety of plots for data visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
#sample Dataframe
data ={
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200,250,300,280,320],
    'Profit' : [50,70,90,85,95]
}
df = pd.DataFrame(data)
```

```python
#Line Plot
df.plot(x='Month', y=['Sales', 'Profit'], kind='line')
plt.title("Monthly Sales and Profit")
plt.show()
```

```python
# Bar plot
df.plot(x='Month', y='Sales', kind='bar', color='skyblue')
plt.title("Monthly sales - Bar Plot")
plt.show()
```

```python
#horizontal bar plot
df.plot(x='Month', y='Sales',kind='barh',color='orange')
plt.title("Monthly sales = Horizontal Bar")
plt.show()
```

```python
# Histogram
df['Sales'].plot(kind='hist',bins=5, color = 'purple')
plt.title("Sales Distribution - Histogram")
plt.show()
```

```python
# Box plot
df[['Sales', 'Profit']].plot(kind='box')
plt.title("Box plot of Sales and Profit")
plt.show()
```

```python
#Area plot
df.set_index('Month')[['Sales','Profit']].plot(kind='area',alpha=0.5)
plt.title("Sales and Profit- Area Plot")
plt.show()
```

```python
#Scatter plot
df.plot(kind='scatter', x='Sales', y='Profit', color= 'green')
plt.title("Sales vs Profit- Scatter Plot")
plt.show()
```

```python
#Pie Chart
df.set_index('Month')['Sales'].plot(kind='pie',autopct='%1.1f%%')
plt.title("Sales Distribution - Pie chart")
plt.ylabel('') #hide y-axis lable
plt.show()
```

```python
#KDE Plot(Kernel Density Estimate)
df['Sales'].plot(kind='kde', color='brown')
plt.title("Sales Density - KDE plot")
plt.show()
```

```python
#Hexbin Plot
# create more data for better hexbin visualization
import numpy as np
df2= pd.DataFrame({
    'X': np.random.randn(1000),
    'Y': np.random.randn(1000) + 1
})
df2.plot(kind= 'hexbin', x='X', y='Y', gridsize=30, cmap='Blues')
plt.title("Hexbin Plot of Random Data")
plt.show()
```
