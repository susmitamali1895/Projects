```python
num_cols = df.select_dtypes(include=["float64", "int64"]).columns
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Categorical Distributions
cat_cols = df.select_dtypes(include=["object"]).columns
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f"Count of {col}")
    plt.show()
```

```python
# correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Box plots
plt.figure(figsize=(6,4))
sns.boxplot(x="Market_Cap_Category", y="Mar Cap - Crore", data=df)
plt.title("Market Cap by Category")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="Sales_Qrt_Category", y="Sales Qtr - Crore", data=df)
plt.title("Quarterly Sales by Category")
plt.show()
```

```python
print("Top 10 Companies by Market Cap")
print(df.nlargest(10, "Mar Cap - Crore")[["Name", "Mar Cap - Crore"]])

print("Bottom 10 Companies by Market Cap")
print(df.nsmallest(10, "Mar Cap - Crore")[["Name", "Mar Cap - Crore"]])

print("Top 10 Companies by Sales")
print(df.nlargest(10, "Sales Qtr - Crore")[["Name", "Sales Qtr - Crore"]])

print("Bottom 10 Companies by Sales")
print(df.nsmallest(10, "Sales Qtr - Crore")[["Name", "Sales Qtr - Crore"]])
```
