```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/ecommerce_customer_churn_dataset.csv")
df
```

```python
df.head()
df.tail()
df.info()
df.shape
df.describe()
```

```python
df['Age']= df['Age'].clip(lower=18, upper=90)
#We will correct erroneous 'Age' values by capping them between 18 and 90, as values outside this range are unrealistic.
df['Cart_Abandonment_Rate']= df['Cart_Abandonment_Rate'].clip(upper=100)
#Sometimes in datasets the calculated value for cart abandonment rate can accidentally go above 100%, which is impossible in real life.
df['Total_Purchases']= df['Total_Purchases'].clip(lower=0)
print("Erroneous values corrected for Age, Cart_Abandonment_Rate and Total_Purchases")
```

```python

num_cols = df.select_dtypes(include= ['int64','float64']).columns


for col in num_cols:
  if col != 'Payment_Method_Diversity': #'Payment_Method_Diversity' will be handled seperately
    df[col].fillna(df[col].median(), inplace=True)
  print("Missing values in numerical columns (except 'Payment_Method_Diversity') imputed with their median")


```

```python
num_cols = df.select_dtypes(include= ['int64','float64']).columns


for col in num_cols:
  if col != 'Payment_Method_Diversity': #'Payment_Method_Diversity' will be handled seperately
    df[col]= df[col].fillna(df[col].median())
  print("Missing values in numerical columns (except 'Payment_Method_Diversity') imputed with their median")
```

```python
df['Payment_Method_Diversity'].fillna(df['Payment_Method_Diversity'].mode()[0], inplace=True)
print("Missing value in 'Payment_Method_Diversity' imputed with its mode.")
```
